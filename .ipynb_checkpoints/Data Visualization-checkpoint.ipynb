```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('Iris.csv')
plt.plot(data["sepal.length"],"r--")
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('Iris.csv')
plt.scatter(x=data['sepal.length'],y=data['petal.length'])
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('Iris.csv')
plt.figure(figsize=(10,5))
x=data["sepal.length"]
plt.hist(x,bins=20,color="red")
plt.title("Sepal length in cm")
plt.xlabel("Sepal length in cm")
plt.ylabel("count")
plt.show()
```
